```python
# importing libraries 
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
```

### Function to Scrape Tweets 
`def scrape_music_tweets(query, start_date, end_date, lang)`
- `query`: keywords used 
- `start_date`: beginning of search 
- `end_date`: ending of search 
- `lang`: language of tweets 

attributes used: 
- `id`: id of tweet 
- `user.username`: username of tweet
- `rawContent`: text content of tweet
- `date`: date tweet was created
- `lang`: language of tweet 
- `outlinks`: url(s) 
- `user.id`: id of user
- `user.followersCount`: user follower count
- `user.friendsCount`: user friend count 
- `user.location`: location of user
- `source`: where tweet was posted from 
- `retweetCount`: retweet count
- `replyCount`: reply count 
- `likeCount`: like count 

```python
def scrape_music_tweets(query, start_date, end_date, lang='en'):
    # create a list to store tweet data
    music_tweets = []

    # use TwitterSearchScraper to scrape data and append tweets to list
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'{query} since:{start_date} until:{end_date} lang:{lang}').get_items()): 
        music_tweets.append([
            tweet.id, # id of tweet 
            tweet.user.username, # username of tweet 
            tweet.rawContent, # text content of tweet
            tweet.date, # date tweet was created
            tweet.lang, # language of tweet 
            tweet.outlinks, # url(s)
            tweet.user.id, # id of user 
            tweet.user.followersCount, # num of followers user has 
            tweet.user.friendsCount, # num of friends user has 
            tweet.user.location, # location of user 
            tweet.source, # where tweet was posted from 
            tweet.retweetCount, # count of retweets
            tweet.replyCount, # count of replies
            tweet.likeCount, # count of likes 
        ])

    return music_tweets
```

### Function to Create Dataframe
`def music_def(music_tweets)`
- `music_tweets`:


```python
def music_df(music_tweets):
    # define the column names for the DataFrame
    column_names = [
        'id_str', 
        'username', 
        'text', 
        'date', 
        'user_language', 
        'URLs', 
        'user_id',
        'followers_count',
        'friends_count',
        'user_location',
        'source_device',
        'retweet_count',
        'reply_count',
        'like_count',
    ]

    # create a DataFrame from the tweet list
    df = pd.DataFrame(music_tweets, columns=column_names)
    df.drop_duplicates(subset=['username'], inplace=True)

    return df
```

### Function to Remove Special/Misc Characters
`def clean_text(text)` 
- `text`:


```python
# removing special characters from 'text' column for sentiment analyses 
import re

# links + special chars
link_pattern = re.compile(r'http\S+')
special_char = re.compile(r'[^\w\s\n]+')

def clean_text(text):
    # remove links
    text = re.sub(link_pattern, '', text)
    # remove special characters
    text = re.sub(special_char, '', text)
    text = re.sub(r'\s+', ' ', text)
    # convert to lowercase
    text = text.lower()
    return text
```

### Extracting Spotify URLs

```python
def extract_spotify_urls(urls):
    return [url for url in urls if 'https://open.spotify.com/track' in url or 'https://spoti.fi/' in url]
```

### Extracting Tracks IDs from Spotify Links

```python
import requests
import re

def extract_track_ids(df):
    # Create an empty list to store the track IDs
    track_ids = []

    # Iterate through each row in the data frame
    for i in range(df.shape[0]):
        row = df.iloc[i]
        # Iterate through each item in the list in the current row
        for link in row['spotify_URLs']:
            if "open.spotify.com/track/" in link:
                # Split the link on '/' and extract the track ID
                track_id = link.split('/')[-1]
                track_ids.append(track_id)
            elif "spoti.fi" in link:
                # Follow the redirection to the actual URL
                r = requests.get(link)
                if r.status_code == 200:
                    # Extract the track ID from the actual URL
                    pattern = r"/track/([a-zA-Z0-9]+)"
                    match = re.search(pattern, r.url)
                    if match:
                        track_id = match.group(1)
                        track_ids.append(track_id)

    # Create a new DataFrame with the unique track IDs
    tracks_df = pd.DataFrame(track_ids, columns=['track_ids'])
    tracks_df.drop_duplicates(subset=['track_ids'], keep='first', inplace=True)

    return tracks_df
```

```python
# getting audio features from spotify urls
def get_audio_features(df, sp):
    # Create an empty list to hold the audio features
    audio_features = []

    # Loop through the rows of the DataFrame and get the audio features for each track
    for index, row in df.iterrows():
        track_id = row['track_ids'].split('?')[0]
        uri = f'spotify:track:{track_id}'
        features = sp.audio_features(uri)[0]
        if features:
            audio_features.append(features)

    return audio_features
```

### Scraping Music Artists from Billboard 

```python
# importing necessary libraries 
import requests
from bs4 import BeautifulSoup
```

```python
def get_billboard_artists(url, num_artists):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    artist_h3s = soup.find_all('h3', class_='c-title')
    artists = ['#' + element.get_text().strip().replace(' ', '') for element in artist_h3s][:num_artists]
    return artists
```

### Getting Track URIs

```python
def get_track_uri(df, sp):
    for i in range(df.shape[0]):
        try:
            track_id = df.loc[i, 'spotify_URLs'].split('/')[-1]
            track_info = sp.track(track_id)
            track_uri = track_info['uri']
            df.at[i, 'track_uri'] = track_uri
        except:
            df.at[i, 'track_uri'] = None
    return df
```

```python
billboard100_df = pd.read_csv('2022-billboard-t100-songs.csv')
billboard100_df = pd.DataFrame(billboard100_df['spotify_URLs'])

billboard100_df['spotify_URLs'] = billboard100_df['spotify_URLs'].astype(str)

billboard100_df = get_track_uri(billboard100_df, sp)
billboard100_df
```

### Getting Audio Features from Track URIs

```python
def get_audio_features(uris):
    # Initialize list to store audio features for each track
    audio_features = []

    # Loop through each URI and retrieve audio features using the Spotify API
    for uri in uris:
        audio_features.append(sp.audio_features(uri)[0])

    # Convert audio features list to a Pandas DataFrame
    audio_features_df = pd.DataFrame(audio_features)

    return audio_features_df
```

```python
billboard100_uris = billboard100_df['track_uri']
billboard100_features = get_audio_features(billboard100_uris)
billboard100_features
```
